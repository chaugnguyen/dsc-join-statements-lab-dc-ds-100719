## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
#Your code here
df = pd.read_sql_query('''SELECT e.firstName, e.lastName
                        FROM employees e
                        JOIN offices o ON e.officeCode = o.officeCode
                        WHERE o.city == "Boston"''', conn)
df
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
#Your code here
df = pd.read_sql_query('''SELECT o.city, count(*) as num_employees
                        FROM employees e
                        JOIN offices o ON e.officeCode = o.officeCode
                        GROUP BY o.city
                        HAVING num_employees = 0''', conn)
df
```

```python
# Your code here
# What is the best-selling product based on quantity and dollars
df = pd.read_sql_query('''SELECT p.productName, 
                        sum(o.quantityOrdered*o.priceEach) as total 
                        FROM products p
                        JOIN orderDetails o ON p.productCode = o.productCode
                        GROUP BY productName
                        ORDER BY total DESC
                        LIMIT 1
                        ''', conn)
df
```

```python
# What is the max/avg/min price product was sold fo 
df = pd.read_sql_query('''SELECT p.productName, p.productCode, sum(o.priceEach)/count(o.priceEach) as avg_price
                        , max(o.priceEach) as max_price, min(o.priceEach) as min_price
                        FROM products p
                        JOIN orderDetails o ON p.productCode = o.productCode
                        GROUP BY p.productName
                        ORDER BY max_price DESC
                        ''', conn)
df

df = pd.read_sql_query('''SELECT p.productName, cast(o.priceEach as float) as pr
                        FROM products p
                        JOIN orderDetails o ON p.productCode = o.productCode
                        WHERE p.productName = '1996 Moto Guzzi 1100i'
                        order by pr
                        
                        ''', conn)
df
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
df = pd.read_sql_query('''SELECT e.firstName, e.lastName, p.productName, sum(od.quantityOrdered) as qty
                        FROM products p
                        JOIN orderDetails od ON od.productCode = p.productCode
                        JOIN orders o ON o.orderNumber = od.orderNumber
                        JOIN customers c ON c.customerNumber = o.customerNumber
                        JOIN employees e ON e.employeeNumber = c.salesRepEmployeeNumber
                        GROUP BY e.firstName, e.lastName, p.productName
                        ORDER BY e.lastName, e.firstName
                        ''', conn)
df

df = pd.read_sql_query('''SELECT e.firstName, e.lastName, count(p.productName) products_sold, sum(od.quantityOrdered) as total_qty
                        FROM products p
                        JOIN orderDetails od ON od.productCode = p.productCode
                        JOIN orders o ON o.orderNumber = od.orderNumber
                        JOIN customers c ON c.customerNumber = o.customerNumber
                        JOIN employees e ON e.employeeNumber = c.salesRepEmployeeNumber
                        GROUP BY e.firstName, e.lastName
                        ORDER BY e.lastName, e.firstName
                        ''', conn)
df

# df = pd.read_sql_query('''SELECT e.firstName, e.lastName, p.productName, od.quantityOrdered, od.orderNumber
#                         FROM products p
#                         JOIN orderDetails od ON od.productCode = p.productCode
#                         JOIN orders o ON o.orderNumber = od.orderNumber
#                         JOIN customers c ON c.customerNumber = o.customerNumber
#                         JOIN employees e ON e.employeeNumber = c.salesRepEmployeeNumber
#                         WHERE productName ='1903 Ford Model A'
#                         AND firstName = 'Loui'
#                         ''', conn)
# df

```
